### **`Variables`**



- Variables are containers for `storing data values`.
- These are entities of a program that `hold a value`.

```python
x = 100
```

- Here `x` is a variable. `100` is the value that is stored in the variable `x`.

### **`Rules for assinging Variables names`**

- A variable name must start with a letter or the underscore character.

```python
x = 100 
```

```python
_x = 100 
```

```python
3x = 100 # error because variables assinment is not possible like this in python
```

```python
@x = 100 # error because variables assinment is not possible like this in python
```

- Variables are case sensitive.

```python
a100 = 10
A100 = 11
```

```python
print(a100)
print(A100)
```

- Reserved words cannot be used as variable names.

```python
break = 10
```

### **Data Types**

Data type represent the type of data stored in a variable.

- `int` - integer
- `float` - floating point number
- `string` - sequence of characters
- `boolean` - true or false
- `complex` - number with a real and imaginary part
- `None` - absence of value

Now let's check one by one.

```python
a = 10
type(a)
```

```python
b = 3.14
type(b)
```

```python
c = "simplilearn"
type(c)
```

```python
d = True
type(d)
```

```python
e = 1 + 2j
type(e)
```

```python
f = None
type(f)
```

### **Some common Data Structures**

- `List` - collection of items in a particular order
- `Tuple` - collection of items in a particular order
- `Set` - collection of items in a particular order
- `Dictionary` - collection of key-value pairs

```python
a = [2,4,6]
print(a)
a[0] = 1 # list is mutable,indexing starts from 0, allows duplicates
print(a)
type(a)
```

```python
a = (2,4,6,6)
print(a)
# a[0] = 1 # tuple is imutable, allows duplicates
print(a[0])  # indexing starts from 0
type(a)
```

### **`Arthematic Operations`**

```python
x = 100
y = 200
```

```python
result = x + y # Addition
print(result)
```

```python
result = x - y # Subtraction
print(result)
```

```python
result = x * y # Multiplication
print(result)
```

```python
result = x / y # Division(result in float)
print(result)
```

```python
result = x // y # Division(result in int)
print(result)
```

```python
result = x % y # Modulo
print(result)
```

```python
result = x ** y # Exponent
print(result)
```
