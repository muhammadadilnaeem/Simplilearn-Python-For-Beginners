### **`Arrays`**

```python
from array import *
```

```python
arr = array('i', [1, 2, 3, 4, 5])
print(arr)
```

```python
print(arr.buffer_info()) # address of the array and length
```

```python
for i in arr: # using for loop to print the array
  print(i)
```

```python
for pnt in range(1,5):
  print(pnt, arr[pnt])
```

```python
print(arr.reverse())
```

```python
arr = array('i', [1, 2,2, 3, 4, 5])
arr.remove(2)
print(arr)
```

```python
arr = array('i',[])
print(arr)
```

```python
# this code will take input from user and store it in array and print it

from array import *
arr = array('i', [])
x = int(input("Enter the length of array: "))
print("Enter %d elements"%x)
for i in range(x):
  y = int(input())
  arr.append(y)
print(arr)
```
