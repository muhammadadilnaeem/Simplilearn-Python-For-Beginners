## **`OOPs`**

```python

# A simple Class

class Car:
  def getSpeed(self):
    print("155 mph.")

BMW = Car()
# Car.getSpeed(BMW)
BMW.getSpeed()

FORD = Car()
# Car.getSpeed(FORD)
FORD.getSpeed()
```

```python
# A class with Constructer

class Car:
  
  def __init__(self,year,speed):
    self.year = year
    self.speed = speed

  def getSpeed(self):
    print(f"Maximum Speed is {self.speed}.")

BMW = Car(2018,155)
# Car.getSpeed(BMW)
BMW.getSpeed()

FORD = Car(2016,140)
# Car.getSpeed(FORD)
FORD.getSpeed()
```

```python
# A class with Constructer and encapsulation

class Car:
  
  def __init__(self,year,speed):
    self.year = year
    self.speed = speed

  def getSpeed(self):
    print(f"Maximum Speed is {self.speed}.")

  def setSpeed(self, speed):
    self.speed = speed

BMW = Car(2018,155)
# Car.getSpeed(BMW)
BMW.setSpeed(169)
BMW.getSpeed()

FORD = Car(2016,140)
# Car.getSpeed(FORD)
FORD.setSpeed(159)
FORD.getSpeed()
```

```python
# A class with Constructer and encapsulation and inheritance

class Car:
  
  def __init__(self,year,speed):
    self.year = year
    self.speed = speed

  def getSpeed(self):
    print(f"Maximum Speed is {self.speed}.")

  def setSpeed(self, speed):
    self.speed = speed

class Sedan(Car):  # Child Class (Inheritance)
  def accelerate(self):
    print('137')
  def opentrunk(self):
    print("Trunk Has Been Opened")

class SUV(Car):
  def accelerate(self):
    print('127')

BMW = Car(2018,155)
FORD = Car(2016,140)
HONDA = Sedan(2018,150)

BMW.getSpeed()
HONDA.getSpeed()
HONDA.opentrunk()
```

```python
class Car: # Parent Class
  def __init__(self,name):
    self.name = name

class Sedan(Car) :  # Child Class
  def accelerate(self):
    print("150")

class SUV(Car):   # Child Class
  def accelerate(self):
    print("180")

objL = [Sedan("Camry"),SUV("Scorpio")]

for obj in objL:
  print(obj.name + " : ",end="")
  obj.accelerate()
```
