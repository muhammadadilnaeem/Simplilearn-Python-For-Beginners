## **`OOPs`**

```python

# A simple Class

class Car:
  def getSpeed(self):
    print("155 mph.")

BMW = Car()
# Car.getSpeed(BMW)
BMW.getSpeed()

FORD = Car()
# Car.getSpeed(FORD)
FORD.getSpeed()
```

```python
# A class with Constructer

class Car:
  
  def __init__(self,year,speed):
    self.year = year
    self.speed = speed

  def getSpeed(self):
    print(f"Maximum Speed is {self.speed}.")

BMW = Car(2018,155)
# Car.getSpeed(BMW)
BMW.getSpeed()

FORD = Car(2016,140)
# Car.getSpeed(FORD)
FORD.getSpeed()
```

```python
# A class with Constructer and encapsulation

class Car:
  
  def __init__(self,year,speed):
    self.year = year
    self.speed = speed

  def getSpeed(self):
    print(f"Maximum Speed is {self.speed}.")

  def setSpeed(self, speed):
    self.speed = speed

BMW = Car(2018,155)
# Car.getSpeed(BMW)
BMW.setSpeed(169)
BMW.getSpeed()

FORD = Car(2016,140)
# Car.getSpeed(FORD)
FORD.setSpeed(159)
FORD.getSpeed()
```
