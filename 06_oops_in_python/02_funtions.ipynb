### **`Functions in Python`**

```python
def welcome():
  print("Good Morning!")

welcome()
```

```python
# Write a Function that can take user name and say welcome {name}

def welcome(name):
  print(f"Welcome {name} !")

name = input('Please Enter Your Name: ')
welcome(name)
```

```python
# this program will give us the the sum of all elements 

def add(*a):
  total = 0
  for i in a:
    total = total + i
  print(f'The sum is {total}.')

add(2,3,4,5,6)
```

```python
# program that can take 2 numbers from a user and return sum

def add(a,b):
  total = float(a + b)
  print(f"The sum of {a} and {b} is {total}.")

a = float(input('Enter First Number: '))
b = float(input('Enter Second Number: '))
add(a,b)
```

```python
def func(lst):
  lst[2] = 0

lst = [0,1,2]
print(f"Before Applying Function = {lst}.")
func(lst)
print(f"After Applying Function = {lst}.")
              
```
