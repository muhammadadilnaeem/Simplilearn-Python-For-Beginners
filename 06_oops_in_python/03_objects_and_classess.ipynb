```python
class Person:
  def __init__(self):
    self.name = "Adil"
    self.gender = "Male"
    self.age = 22

  def talk(self):
    print(f"Hi! I am {self.name}.")

  def vote(self):
    if self.age >=18:
      print("I am Eligible and I Can Vote.")
    else:
      print("I am not Eligible and I Can not Vote.")

obj = Person()
Person.talk(obj)
Person.vote(obj)
```

```python

# If we want to keep unique values for every object we can do this 

class Person:
  def __init__(self,n,g,a):
    self.name = n
    self.gender = g
    self.age = a

  def talk(self):
    print(f"Hi! I am {self.name}.")

  def vote(self):
    if self.age >=18:
      print("I am Eligible and I Can Vote.")
    else:
      print("I am not Eligible and I Can not Vote.")

obj1 = Person('Adil', "Male", 21)
obj2 = Person('Ali', "Female", 22)

print(obj1.name,obj2.name)
print(obj1.age,obj2.age)
```
