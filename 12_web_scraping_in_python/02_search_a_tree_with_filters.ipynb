### **`Web Scraping in Python`**

##### **Import Libraries**

```python
from bs4 import BeautifulSoup
```

##### **Import Web Scrapping Example Html file**

```python
HTMLfilePATH = 'web_scrapping_example.html'
with open(HTMLfilePATH,"r") as organization:
  soup = BeautifulSoup(organization, 'lxml')
```

##### **View Content of Soup Object**

```python
soup.contents
```

##### **Search Using Find Methods**

```python
tag_list = soup.find('li')
```

```python
print(tag_list)
print(type(tag_list))
```

##### **Search the document using find method for an ID**

```python
find_id = soup.find(id='HR')

# Print find id object
print(find_id)
```

##### **Search using String only**

```python
search_for_string_only = soup.findAll(text=['Kelly','Jack'])

# Print the search values
print(search_for_string_only)
```

##### **Search on basse of CSS class only**

```python
css_class_search = soup.find(attrs={'class':'HRmanager'})
print(css_class_search)
```

##### **Use Reguler Expressions to Search the document**

```python
# Import necessary libraries
import re
from bs4 import BeautifulSoup

# Example HTML email content
email_example = """
<html>
<body>
    <div>Email ID: example@example.com</div>
</body>
</html>
"""

# Parse the email example
soup = BeautifulSoup(email_example, "lxml")

# Compile the regular expression to find email IDs
emailid_regex = re.compile(r'\w+@\w+\.\w+')

# Find and print the email ID using regular expression
email_id = soup.find(text=emailid_regex)
print(email_id)
```
