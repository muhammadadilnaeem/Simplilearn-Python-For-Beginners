## **`Web Scraping in Python`**

#### **Import Libraries**

```python
import os
import bs4
from bs4 import BeautifulSoup
```

#### **Create an HTML document**

```python
html_doc = """
<html>
  <body>
    <h1>My First Heading</h1>
    <b>This is a comment line</b>
    <p title="About Me" class="test">My first paragraph.</p>
    <div class="cities">
      <h2>London</h2>
    </div>
  </body>
</html>
"""
```

#### **Parse it using the HTML parser**

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

#### **Print the parsed HTML**

```python
print(soup.prettify())
```

#### **Print the type of soup**

```python
print(type(soup))
```

#### **Create a Tag Object**

```python
tag = soup.p
```

#### **View the tag object type**

```python
type(tag)
```

#### **Print the tag object**

```python
print(tag)
```

#### **View tag Attribute**

```python
tag.attrs
```

#### **View tag attribute**

```python
tag.string
```

```python
type(tag.string)
```
