```python
i = 20

if i % 2 == 0:
    print("This is an If block.")
    print("'i' is an Even number.")
else:
    print("This is an else block.")
    print("'i' is an Odd number.")
```

```python
c = 25
if c < 25:
    if c % 2 == 0:
        print("This is an If if block.")
        print("'c' is an Even number.")
    else:
        print("This is an if else block.")
        print("'c' is an Odd number.")
else:
    print("This is an else else block.")
    print("'c' is greater than or equal to 25.")
```

```python
var = "z"

if var=="a":
    print("This is a vovel 'a'.")
elif var=="e":
    print("This is a vovel 'e'.")
elif var=="i":
    print("This is a vovel 'i'.")
elif var=="o":
    print("This is a vovel 'o'.")
elif var=="u":
    print("This is a vovel 'u'.")
else:
    print("This is not a vovel. It is a consonant.")
```

```python
var = 'z'

if var == 'a':
    print("This is an If block.")
    print("'var' is an 'a'.")
elif var == 'b':
    print("This is an elif block.")
    print("'var' is not 'b'.")
else:
    print("This is an else block.")
    print("'var' is not 'a' or 'b'.")
    
```
