## **`Threading in Python`**


- Threading is used for running multiple tasks simultaneously.

```python
from threading import *

def show():
  print("This is a Child Thread. ")

t = Thread(target=show)
t.start()
print("This is a Parent Thread.")
```

```python
from threading import *

class MyThread(Thread):
  def run(self):
    for i in range(5):
      print("This is Child Thread.")

t = MyThread()
t.start()
for i in range(5):
  print("\nThis is Parent Thread.")
```

```python
from threading import *
class Demo:
  def show(self):
    for i in range(5):
      print("This is Child Thread.")

obj = Demo()
t = Thread(target=obj.show)
t.start()
for i in range(5):
  print("\nThis is Parent Thread.")
```

```python
from threading import *

class Demo:
  def num(self):
    for i in range(1,6):
      print(f"The number is {i}")
  def double(self):
    for i in range(1,6):
      print(f"The double number is {i*2}")
  def square(self):
    for i in range(1,6):
      print(f"The square number is {i**2}")

obj = Demo()
t1 = Thread(target=obj.num)
t2 = Thread(target=obj.double)
t3 = Thread(target=obj.square)

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()

print("This is the Main Thread")
```
