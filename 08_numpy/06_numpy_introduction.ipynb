##### **How much memory a List Uses**

```python
b = range(1000)
print(sys.getsizeof(5)*len(b)) # this will print the size of the list
```

##### **How much memory a array Uses**

```python
c = np.arange(1000)
print(c.size*c.itemsize) # this will print the size of the array
```

##### **How much speed a numpy array Uses to perform a task**

```python
size = 100000
```

```python
L1 = range(size)
L2 = range(size)
A1 = np.arange(size)
A2 = np.arange(size)
```

```python
# for list

start = time.time()
result = [(x+y) for x,y in zip(L1,L2)]
print("python list took: ", (time.time()-start)*10000, "ms")
```

```python
# for array

start = time.time()
result = A1+A2
print("numpy array took: ", (time.time()-start)*10000, "ms")
```
