### **`Scipy in Python`**

```python
from scipy import constants
```

```python
print(constants.pi)  # pi
print(constants.e)   # e
print(constants.c)   # speed of light
print(constants.N_A) # Avogadro's number
print(constants.h)   # Planck's constant
```
