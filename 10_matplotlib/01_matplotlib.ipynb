### **`Matplotlib in Python`**

```python
# Importing Libraries

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a list of evenly-spaced numbers over a range

x = np.linspace(0, 20, 100)  # Create a list of evenly-spaced numbers over the range
plt.plot(x, np.sin(x))       # Plot the sine of each x point
plt.show()                   # Display
```

```python
# Radar chart

fig = plt.figure(figsize=(8,8))
ax = fig.add_axes([0.0,0.0,1,1], polar=True)
t = np.linspace(0, 2*np.pi, 100)
ax.plot(t,t*2, color='red', linewidth=3)
```

```python
# Draw a histogram

n = np.random.randn(100000)
fig, ax = plt.subplots(1,2, figsize=(12,4))

ax[0].hist(n)
ax[1].hist(n, bins=50)

plt.show()
```

```python
# Draw a counter map
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)   # Grid of point
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
fig,ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=True, fontsize=10)
plt.show()
```

```python
plt.imshow(Z, interpolation='bilinear', cmap=cm.gist_earth,
              origin='lower', extent=[-3, 3, -2, 2])
plt.show()
```
