## **`Strings in Python`**

```python
stg = 'tim' # single quote
print(stg)
print(type(stg))
```

```python
stg1 = 'tim's birthday' # error due to single quote
```

```python
stg2 = "tim's birthday" # double quote
print(stg2)
print(type(stg2))
```

#### **`Built-in Functions with Strings`**

```python
a = 'Hello there My name is Adil I am from Lahore, Pakistan.' # string is a collection of characters
len(a)
```

```python
print(a[6:11])
```

```python
for i in a: # for loop to print each character of the string
   print(i)
```

```python
print(a.upper()) # to convert all characters in the string to upper case
```

```python
print(a.lower()) # to convert all characters in the string to lower case
```

```python
a = "Welcome to Simplilearn, I am from Lahore, Pakistan"
```

```python
print(a.index('P')) # to find the index of a character
```

```python
print(a.split(","))
```

```python
a.replace("Lahore", "Karachi") # to replace a character
```

```python
s1 = "hello"
s2 = "world"
s3 = s1 + " " + s2 # to concatenate two strings
print(s3)
```
