## **`Sets in Python`**

```python
s = set([1, 2, 3, 4])
print(s)
print(type(s))
```

```python
s.add(5) # add an element
print(s)
```

```python
fs = frozenset([1, 2, 3, 4])
print(fs)
```

```python
fs.add(5) # error because frozenset is immutable
```

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
s1.union(s2)   # union of sets
```

```python
s1.intersection(s2)  # intersection of sets
```

```python
s1.difference(s2)  # difference of sets
```
