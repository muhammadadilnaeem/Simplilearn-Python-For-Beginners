```python
# Write a program that lets the user enter a multiple of 7

var = int(input("Enter a multiple of 7: "))

while var % 7 != 0:
    var = int(input("Enter a multiple of 7: "))

print(f"Number {var} is a multiple of 7")
```

```python
# Accessing the elements of a matrix

# for in for loop

x = [[1,2,3],['a','b','c']]

for i in x:
    for j in i:
        print(j, end='') # end='' removes the new line
    print() # prints a new line
```

```python
# use of break statement

x = 'Hello there, how are you?'

for i in x:
    if i == ',':
        break
    print(i, end='')
```

```python
# use of continue statement

x = 'Hello there, how are you?'

for i in x:
    if i == ',':
        continue
    print(i, end='')
```
