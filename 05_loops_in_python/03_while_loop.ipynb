### **`While Loop in Python`**


```python
# print 10 times 'Simplilearn' using while loop
i = 1
while i < 11:
  print('Simplilearn')
  i +=1
```

```python
# print 10 times 'Simplilearn' using while loop
i = 11
while i > 1:
  print('Simplilearn')
  i -=1
```

```python
# A program to print 'simplilearn' n number of times using while loop 

n = int(input('Enter number: '))
while n > 0:
  print('simplilearn', end=' ')
  n = n - 1
```

```python
# find sum of first 10 natural numbers using while loop
i = 1
sum = 0
while i <= 10:
  sum = sum + i
  i = i + 1
print(sum)
```

```python
i = 1
sum = 0

while i <= 10:
    if i % 2 == 0:
        sum = sum + i
    i = i + 1

print(sum)
```

```python
# write a program to calculate length of a list without using len() function

x = [1, 2.3, 'simplilearn']
length = 0
i = 0
try:
    while x[i] :
      length = length + 1
      i = i + 1
except IndexError:
  print(length)
```

```python
# print a pattren of same number using while loop
n = int(input('Enter number: '))
i = 1
while i <= n:
  j=1
  while j <= i:
    print(i, end=' ')
    j = j + 1
  i = i + 1
  print()
```

```python
import random

# Generate a random 4-digit number
nump = random.randint(1000, 9999)
print("A 4-digit number has been generated. Try to guess it!")
print("Type 'quit' to exit the game.")

while True:
    user_input = input('Enter a 4-digit number (or type "quit" to exit): ')
    
    if user_input.lower() == 'quit':
        print('Thanks for playing! Goodbye!')
        break

    # Ensure the input is a 4-digit number
    if not user_input.isdigit() or len(user_input) != 4:
        print("Please enter a valid 4-digit number.")
        continue

    n = int(user_input)

    if n == nump:
        print('Congrats! You guessed it right')
        break

    cor = 0
    temp_nump = nump
    temp_n = n

    # Count the number of correct digits
    for _ in range(4):
        numc = temp_nump % 10
        nc = temp_n % 10
        if numc == nc:
            cor += 1
        temp_nump //= 10
        temp_n //= 10

    if cor == 4:
        print('Congrats! You guessed it right')
        break
    else:
        print('%d digits are correct' % cor)
```
